```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Carregando os dados

df_train = pd.read_csv('train.csv')
df_test = pd.read_csv('test.csv')

# 2. Verificando as dimensões dos dados
print(f"Tamanho da base de treino: {df_train.shape[0]} linhas e {df_train.shape[1]} colunas")
print(f"Tamanho da base de teste: {df_test.shape[0]} linhas e {df_test.shape[1]} colunas")

# 3. Listando as 5 primeiras linhas 
display(df_train.head(10))

# 4. Informações gerais (tipos de dados e contagem de não-nulos)
print("\n--- Informações do Dataset de Treino ---")
df_train.info()

# 5. Estatísticas descritivas iniciais das variáveis numéricas
display(df_train.describe())

# 6. Verificando a distribuição da nossa variável alvo (Target)
print("\n--- Distribuição do Score de Crédito (%) ---")
print(df_train['Credit_Score'].value_counts(normalize=True) * 100)
```

```python
# 1. Algumas colunas tiveram os dados númericos classificados como object. Fazendo a correção
colunas_para_limpar = [
    'Age', 'Annual_Income', 'Num_of_Loan', 
    'Num_of_Delayed_Payment', 'Changed_Credit_Limit', 
    'Outstanding_Debt', 'Amount_invested_monthly', 
    'Monthly_Balance'
]

# 2. Iterando as duas bases (treino e teste) para não repetir código
for df in [df_train, df_test]:
    for col in colunas_para_limpar:
        # Passo A: Transformamos em string e removemos o caractere '_' que costuma poluir esses dados
        df[col] = df[col].astype(str).str.replace('_', '')
        
        # Passo B: Forçamos a conversão para numérico (float). 
        # O errors='coerce' garante que qualquer texto puro que sobrar (ex: "NA", "N/A", espaços vazios) 
        # seja transformado em NaN (Not a Number / Valor Nulo), o que é o correto estatisticamente.
        df[col] = pd.to_numeric(df[col], errors='coerce')

# 3. Verificando o resultado no dataframe de treino
print("--- Dtypes corrigidos ---")
print(df_train[colunas_para_limpar].info())

# 4. Agora que transformamos a "sujeira" em valores nulos (NaN), vamos contar quantos nulos temos no total
print("\n--- Contagem de Valores Nulos (Missing Values) por Coluna ---")
display(df_train.isna().sum()[df_train.isna().sum() > 0]) # Mostra apenas colunas que têm nulos
```

```python
# 1. Calcula a porcentagem de nulos para todas as colunas
valores_nulos_pct = (df_train.isna().mean() * 100).round(2)

# 2. Filtra para exibir apenas as colunas que possuem algum valor nulo
print("--- Porcentagem de Valores Nulos por Coluna (%) ---")
print(valores_nulos_pct[valores_nulos_pct > 0].sort_values(ascending=False))
```

```python
import re
# Tratamento da coluna Credit_History_Age
def converter_idade_credito(texto):
    # Se o valor já for nulo, retornamos nulo
    if pd.isna(texto) or not isinstance(texto, str):
        return np.nan
    
    # Busca um ou mais dígitos (\d+) seguidos da palavra 'Year'
    match_anos = re.search(r'(\d+)\s*Year', texto)
    # Busca um ou mais dígitos (\d+) seguidos da palavra 'Month'
    match_meses = re.search(r'(\d+)\s*Month', texto)
    
    anos = int(match_anos.group(1)) if match_anos else 0
    meses = int(match_meses.group(1)) if match_meses else 0
    
    # Calcula o total de meses
    total_meses = (anos * 12) + meses
    
    # Se não encontrou nenhum dos dois, pode ser uma sujeira do tipo "NA"
    if not match_anos and not match_meses:
        return np.nan
        
    return total_meses

# Aplicando a transformação nas duas bases (Treino e Teste)
df_train['Credit_History_Months'] = df_train['Credit_History_Age'].apply(converter_idade_credito)
df_test['Credit_History_Months'] = df_test['Credit_History_Age'].apply(converter_idade_credito)

# Vamos ver como ficou o resultado estruturado
display(df_train[['Credit_History_Age', 'Credit_History_Months']].head(15))
```

```python
import pandas as pd
import numpy as np

# 1. Separamos as colunas em Numéricas e Categóricas (Textos)
colunas_numericas = df_train.select_dtypes(include=['float64', 'int64']).columns
colunas_categoricas = df_train.select_dtypes(include=['object']).columns

# 2. Preenchendo as Numéricas com a MEDIANA
for col in colunas_numericas:
    # Calculamos a mediana usando a base de TREINO (sempre aprendemos do treino)
    mediana = df_train[col].median()
    
    # Preenchemos os nulos no Treino
    df_train[col] = df_train[col].fillna(mediana)
    # E preenchemos no Teste (só se a coluna existir lá)
    if col in df_test.columns:
        df_test[col] = df_test[col].fillna(mediana)

# 3. Preenchendo as Categóricas (Textos) com a MODA (o valor mais frequente)
for col in colunas_categoricas:
    # Ignoramos a Variável Alvo para não dar problema, e colunas irrelevantes como ID
    if col == 'Credit_Score':
        continue # O 'continue' faz o código pular para a próxima coluna
        
    moda = df_train[col].mode()[0]
    
    df_train[col] = df_train[col].fillna(moda)
    
    # Preenchemos no Teste (só se a coluna existir lá)
    if col in df_test.columns:
        df_test[col] = df_test[col].fillna(moda)

# 4. Verificação final: Tem que aparecer "0" (nenhum nulo nas variáveis que usaremos)
print("Quantidade total de valores nulos após a limpeza:")
print(f"Base de Treino: {df_train.isna().sum().sum()} nulos")
print(f"Base de Teste: {df_test.isna().sum().sum()} nulos")
```

```python
import numpy as np

# --- 1. CORRIGINDO OS OUTLIERS DE IDADE ---
for df in [df_train, df_test]:
    # Transforma idades impossíveis (menores que 18 ou maiores que 100) em Nulos (NaN)
    df.loc[(df['Age'] < 18) | (df['Age'] > 100), 'Age'] = np.nan

# Calculamos a nova mediana apenas com as idades normais
nova_mediana_idade = df_train['Age'].median()

# Preenchemos os buracos que acabamos de criar com essa nova mediana
df_train['Age'] = df_train['Age'].fillna(nova_mediana_idade)
df_test['Age'] = df_test['Age'].fillna(nova_mediana_idade)


# --- 2. NOVA ANÁLISE: DÍVIDA PENDENTE vs SCORE ---
plt.figure(figsize=(10, 6))

# Boxplot para ver como a Dívida se comporta em cada categoria de Score
sns.boxplot(data=df_train, x='Credit_Score', y='Outstanding_Debt', palette='Set2')

plt.title('Impacto da Dívida Pendente no Score de Crédito', fontsize=12, pad=15)
plt.xlabel('Score de Crédito', fontsize=11)
plt.ylabel('Dívida Pendente (US$)', fontsize=11)

# Como podem ter dívidas muito altas, vamos limitar o eixo Y para enxergar melhor a "caixa" principal
# Vamos focar em quem deve até 5000 para ver o padrão geral
plt.ylim(0, 5000) 
plt.yticks(np.arange(0, 7001, 500))
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# 1. Escolhendo um conjunto forte de variáveis numéricas para o nosso primeiro modelo
features_selecionadas = [
    'Age', 'Num_Bank_Accounts', 'Num_Credit_Card', 'Interest_Rate',
    'Delay_from_due_date', 'Credit_Utilization_Ratio',
    'Total_EMI_per_month', 'Outstanding_Debt', 'Credit_History_Months'
]

# Garantindo que todas as selecionadas sejam números de verdade (blindagem extra)
for col in features_selecionadas:
    df_train[col] = pd.to_numeric(df_train[col].astype(str).str.replace('_', ''), errors='coerce')

# Preenchendo qualquer buraquinho que tenha sobrado com a mediana
df_train[features_selecionadas] = df_train[features_selecionadas].fillna(df_train[features_selecionadas].median())

# 2. Separando as características do cliente (X) e o alvo que queremos prever (y)
X = df_train[features_selecionadas]
y = df_train['Credit_Score']

# 3. Usando o LabelEncoder para transformar "Good", "Standard", "Poor" em 0, 1, 2
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# 4. Dividindo os dados: 80% para o modelo aprender, 20% para aplicarmos uma prova e vermos se ele aprendeu bem
X_train, X_val, y_train, y_val = train_test_split(X, y_encoded, test_size=0.2, random_state=42)

print(f"Base de treino separada com {X_train.shape[0]} clientes.")
print(f"Base de validação (prova final) separada com {X_val.shape[0]} clientes.")
print("\n--- Dicionário do Score de Crédito ---")

# Mostrando como o modelo vai entender as palavras a partir de agora
for i, classe in enumerate(le.classes_):
    print(f"O número {i} representa o Score '{classe}'")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Criando o "cérebro" do nosso modelo (Agora com Balanceamento de Classes!)
# n_estimators=100 significa que teremos 100 "analistas" (árvores) votando
# random_state=42 garante que sempre teremos o mesmo resultado se rodarmos o código de novo
# class_weight='balanced' avisa o modelo para dar mais atenção às classes com menos clientes
rf_model = RandomForestClassifier(n_estimators=200, random_state=42, n_jobs=-2)

print("Treinando o modelo Random Forest Balanceado")
rf_model.fit(X_train, y_train)

# 3. O modelo faz a "prova": tenta adivinhar o score dos 20% de clientes que ele nunca viu
y_pred = rf_model.predict(X_val)

# 4. Calculando a Acurácia (a porcentagem de vezes que ele acertou na mosca)
acuracia = accuracy_score(y_val, y_pred)

print(f"\n--- Resultado da Prova ---")
print(f"Acurácia Geral do Modelo: {acuracia * 100:.2f}%\n")

# 5. Imprimindo o "Boletim Detalhado" (Classification Report)
print("--- Boletim Detalhado por Categoria ---")
# passamos as 'classes_' do nosso LabelEncoder para aparecer os nomes originais em vez de 0, 1 e 2
print(classification_report(y_val, y_pred, target_names=le.classes_))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# 1. O computador calcula onde o modelo acertou e onde ele errou
cm = confusion_matrix(y_val, y_pred)

# 2. Configurando o tamanho do gráfico
plt.figure(figsize=(7, 5))

# 3. Criando o mapa de calor (Heatmap)
# annot=True coloca os números dentro dos quadrados
# fmt='d' formata como número inteiro
# cmap='Blues' usa tons de azul para colorir (mais escuro = mais clientes)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
            xticklabels=le.classes_, yticklabels=le.classes_)

# 4. Colocando títulos e rótulos para facilitar a leitura
plt.title('Matriz de Confusão - Random Forest', fontsize=12, pad=13)
plt.ylabel('Realidade', fontsize=11, fontweight='regular')
plt.xlabel('Previsão', fontsize=11, fontweight='regular')

plt.show()
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier
from sklearn.metrics import classification_report, accuracy_score

# Usamos o HistGradientBoosting porque ele é ultrarrápido para bases grandes (como a nossa de 100 mil linhas)
# random_state=42 garante que o resultado será o mesmo se você rodar de novo
gb_model = HistGradientBoostingClassifier(random_state=42, max_iter=50000)

print("Treinando o modelo Gradient Boosting... Ele aprende corrigindo os próprios erros!")
gb_model.fit(X_train, y_train)

y_pred_gb = gb_model.predict(X_val)

acuracia_gb = accuracy_score(y_val, y_pred_gb)

print(f"\n--- Resultado da Prova (Gradient Boosting) ---")
print(f"Acurácia Geral do Modelo: {acuracia_gb * 100:.2f}%\n")

print("--- Boletim Detalhado por Categoria ---")
print(classification_report(y_val, y_pred_gb, target_names=le.classes_))
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report

# 1. Configurando o XGBoost com Tuning
xgb_model_tunado = xgb.XGBClassifier(
    objective='multi:softprob', 
    num_class=3, 
    random_state=42,
    eval_metric='mlogloss',
    n_estimators=500,          # Mais árvores
    learning_rate=0.05,        # Passos mais cautelosos
    max_depth=6,               # Limita a profundidade
    subsample=0.8,             # Usa 80% das linhas por vez
    colsample_bytree=0.8       # Usa 80% das colunas por vez
)

# 2. Treinando o modelo (Vai demorar um pouco mais que o normal!)
print("Treinando o XGBoost Tunado (Aguarde alguns segundos)...")
xgb_model_tunado.fit(X_train, y_train)

# 3. Aplicando a prova
y_pred_xgb_tunado = xgb_model_tunado.predict(X_val)

# 4. Avaliando os resultados
acuracia_xgb_tunado = accuracy_score(y_val, y_pred_xgb_tunado)
print("\n--- Resultado da Prova (XGBoost TUNADO) ---")
print(f"Acurácia Geral do Modelo: {acuracia_xgb_tunado * 100:.2f}%\n")
print("--- Boletim Detalhado por Categoria ---")

nomes_classes = ['Good', 'Poor', 'Standard']
print(classification_report(y_val, y_pred_xgb_tunado, target_names=nomes_classes))
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

print("Iniciando a Validação Cruzada (5 Folds)... Isso vai treinar o modelo 5 vezes!")

# 1. O cross_val_score divide a base de treino em 5 pedaços diferentes (cv=5).
# Ele treina em 4 pedaços e testa no 1 que sobrou, repetindo o processo até testar tudo.
# Usamos n_jobs=-1 para usar todos os núcleos do seu computador e ir mais rápido.
notas_cv = cross_val_score(rf_model, X_train, y_train, cv=5, scoring='accuracy', n_jobs=-1)

# 2. Mostrando as 5 notas que o modelo tirou
print(f"\nNotas individuais de cada teste: {notas_cv}")

# 3. Calculando a média e o desvio padrão (para ver a estabilidade)
media_cv = np.mean(notas_cv) * 100
desvio_cv = np.std(notas_cv) * 100

print(f"\n--- Resultado Oficial Consolidado ---")
print(f"Acurácia Média Final: {media_cv:.2f}%")
print(f"Variação (Estabilidade): +/- {desvio_cv:.2f}%")
```

```python
import shap
import matplotlib.pyplot as plt

print("Calculando os valores SHAP... Abrindo a caixa preta do modelo!")

# 1. Reduzimos a amostra de 1000 para 150 clientes. 
# O gráfico vai ser gerado em segundos e a conclusão será a mesma!
X_amostra = X_train.sample(n=150, random_state=42)

# 2. Criando o "Explicador" específico para algoritmos de árvore (Random Forest)
explicador = shap.TreeExplainer(rf_model)

# 3. Calculando o impacto (check_additivity=False desliga uma conta pesada e acelera muito)
valores_shap = explicador.shap_values(X_amostra, check_additivity=False)

# 4. Gerando o Gráfico! 
# Como temos 3 classes (Good, Standard, Poor), o shap_values é uma lista com 3 partes.
# Vamos olhar especificamente como o modelo decide se alguém é "Poor" (que é a classe 1 no nosso LabelEncoder)
plt.figure(figsize=(10, 6))
plt.title('Como as variáveis empurram o cliente para o Score "Poor"?', fontsize=14)

# Gerando o gráfico de "enxame de abelhas" (Beeswarm) para a classe 1 (Poor)
shap.summary_plot(valores_shap[1], X_amostra, show=False)

plt.tight_layout()
plt.show()
```
